```python
import numpy as np
xishui["ma5"] = np.round(xishui["close"].rolling(window = 5, center = False).mean(), 2)
xishui["ma10"] = np.round(xishui["close"].rolling(window = 10, center = False).mean(), 2)
xishui["ma20"] = np.round(xishui["close"].rolling(window = 20, center = False).mean(), 2)
```
